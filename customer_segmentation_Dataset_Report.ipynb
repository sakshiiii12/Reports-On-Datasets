```python
#Income vs. Spending Score
sns.scatterplot(x='Spending_Score', y='Annual_Income_k$', color='blue', data=df)
plt.title("Income VS Spending Score Trends")
plt.show()
```

```python
#Feature Engineering
df['Total_Purchases'] = df['Online_Purchases'] + df['InStore_Purchases']
```

```python
df.head(5)
```

```python
sns.countplot(x='Total_Purchases', data=df, color='green')
plt.title("Total Purchases Trends")
plt.show()
```

```python
df.columns
```

```python
#Segmenting Customers into Age Groups
bins = [17, 25, 35, 50, 70] 
labels = ['18-25', '26-35', '36-50', '51+']
df['Age_Group'] = pd.cut(df['Age'], bins=bins, labels=labels)
df[['Age', 'Age_Group']].head()
```

```python
sns.countplot(data=df, x='Age_Group', order=labels, palette='magma')
plt.title("Customer Count by Age Group")
plt.xlabel("Age Group")
plt.ylabel("Count")
plt.show()
```

```python
#Heatmap for correlation between numeric features
correlation_matrix = df.corr(numeric_only=True)
sns.heatmap(correlation_matrix, annot=True, cmap='cool', fmt=".2f", linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()
```
