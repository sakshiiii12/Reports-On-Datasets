```python
#Data Visulizations
```

```python
#ðŸ“ˆ Line chart: Monthly sales trend.
df['Year-Month'] = df['Order Date'].dt.to_period('M').astype(str)
```

```python
monthly_sales = df.groupby('Year-Month')['Sales'].sum().reset_index()
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=monthly_sales, x='Year-Month', y='Sales', marker='o', color='red')
plt.title("Monthly Sales Trend")
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.tight_layout()
plt.grid(True)
plt.show()
```

```python
#ðŸ“Š Bar plots: Top states by profit and loss.
state_profit = df.groupby('State')['Profit'].sum().reset_index()
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(data=state_profit, x='State', y='Profit', color='pink')
plt.title("Top States By Profit and Loss Trends")
plt.xlabel("State")
plt.ylabel("Profit")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
#ðŸ“‰ Scatter plot: Sales vs Profit.
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Sales', y='Profit', hue='Category')
plt.title('Sales vs Profit by Category')
plt.xlabel('Sales')
plt.ylabel('Profit')
plt.tight_layout()
plt.show()
```

```python
df.columns
```

```python
#ðŸ“¦ Count plot: Orders by shipping mode.
sns.countplot(x= 'Ship Mode', data=df, palette='pastel')
plt.title('Number of Orders by Shipping Mode')
plt.xlabel('Shipping Mode')
plt.ylabel('Order Count')
plt.tight_layout()
plt.show()
```

```python
#ðŸŒ¡ Heatmap: Profit by Region vs Category.
region_category_profit = df.groupby(['Region', 'Category'])['Profit'].sum().reset_index()
```

```python
heatmap_data = region_category_profit.pivot(index='Region', columns='Category', values='Profit')
```

```python
sns.heatmap(heatmap_data, annot=True, fmt=".0f", cmap="YlGnBu")
plt.title('Profit by Region and Category')
plt.xlabel('Category')
plt.ylabel('Region')
plt.tight_layout()
plt.show()
```
